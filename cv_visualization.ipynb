```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
field = cv2.imread('im1.jpg')
field = cv2.cvtColor(field, cv2.COLOR_BGR2RGB)
plt.imshow(field)
plt.show()
```

```python
hsv_field = cv2.cvtColor(field, cv2.COLOR_RGB2HSV)
```

```python
min_color = (0, 100, 100)
max_color = (20, 255, 255)
```

```python
mask = cv2.inRange(hsv_field, min_color, max_color)
```

```python
result = cv2.bitwise_and(field, field, mask=mask)
```

```python
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
plt.imshow(result)
plt.show()
```

```python
field_2 = cv2.imread('im2.png')
field_2 = cv2.cvtColor(field_2, cv2.COLOR_BGR2RGB)
plt.imshow(field_2)
plt.show()
```

```python
plt.imshow(mask_2, cmap="gray")
plt.show()
```

```python
plt.imshow(res)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1, 3, 1)
plt.imshow(field_2)
plt.subplot(1, 3, 2)
plt.imshow(mask_2, cmap="gray")
plt.subplot(1, 3, 3)
plt.imshow(res)
```
